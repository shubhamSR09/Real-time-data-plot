```python
import time
import psutil
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
plt.rcParams['animation.html']='jshtml'
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
fig.show()
```

```python
i = 0
x,y=[],[]
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y,color='b')
    fig.canvas.draw()
    time.sleep(0.5)
    i+=1
```
